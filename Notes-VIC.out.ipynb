## **Energy Efficiency in New Residential Buildings**

### **Policy Goals**

Victoria has implemented stringent energy efficiency requirements for new homes to reduce energy consumption, greenhouse gas emissions, and improve occupant comfort[1]. The state aims to achieve net-zero emissions by 2050 and has set ambitious targets to reduce greenhouse gas emissions by 45-50% from 2005 levels by 2030[2]. Given that residential buildings are a significant contributor to Australia’s overall emissions, they represent a key area for improvement[3].

### **Implementation Strategies**

#### **7-Star Energy Efficiency Standards**

In 2022, Victoria raised the minimum energy efficiency building standards for new homes from 6 to 7 stars[4]. This change necessitates improvements in insulation, window glazing, and the use of energy-efficient appliances in new homes[5]. The 7-star standard ensures that new homes are designed and built to minimize energy consumption and maximize thermal comfort. Notably, this shift towards higher energy efficiency standards has created a demand for new skills and occupations in the building sector, with professionals now required to design and implement these energy-efficient measures[6].

To support the transition to 7-star homes, Sustainability Victoria’s 7 Star Homes program provided rebates and verification assessments to builders and land developers[7]. This program aimed to accelerate the adoption of energy-efficient practices in the residential building market. For instance, the program offered \$4,000 rebates per home, for up to 4 homes built, to assist with upgrade costs and provided as-built verification assessments to ensure homes are built to achieve the same efficiency rating as designed[8].

#### **National Construction Code (NCC)**

The NCC establishes the energy efficiency requirements for new homes, renovations, alterations, and additions[9]. The NCC aims to enhance the sustainability and liveability of buildings by reducing energy consumption and greenhouse gas emissions while improving occupant health and amenity[10].

Performance requirement H6P1 in the NCC mandates that habitable rooms and conditioned spaces use less energy than the calculated heating and cooling loads[11]. This requirement promotes the construction of building fabrics that minimize reliance on artificial heating and cooling, thereby reducing energy consumption.

## **Energy Efficiency in Existing Residential Buildings**

Recognizing that existing homes constitute a substantial portion of the housing stock, Victoria has implemented various programs and initiatives to improve their energy efficiency.

The One Million Homes Alliance, a collaborative effort between leading environment, social sector, local government, and consumer organizations, is committed to transforming the efficiency and affordability of Victoria’s housing[12]. This alliance advocates for policies and programs that promote energy efficiency upgrades in existing homes.

### **Victorian Energy Upgrades (VEU) Program**

The VEU program is a key initiative aimed at reducing greenhouse gas emissions by providing Victorians with access to discounted energy-efficient products and services[13]. The program encourages investment, employment, and innovation in industries that supply these products and services[14].

Under the VEU program, large energy retailers are required to acquire and surrender Victorian energy efficiency certificates (VEECs) to meet annual targets set by legislation[15]. Accredited persons who install energy-efficient products in homes can create VEECs[16].

The VEU program offers various incentives for homeowners to improve their energy efficiency, including discounts on a range of upgrades. These are summarized in the table below:

| Upgrade Type | Current System | New System | Discount | Annual Savings |
|:--------------|:--------------|:--------------|:--------------|:--------------|
| Heating and Cooling | Gas ducted heating system | Electric heating system | Up to \$3,600 | Around \$600 |
| Hot Water System | Gas water heater | Electric heat pump hot water system | N/A | \$250 |

In addition to these upgrades, VEU also offers home energy rating assessments to provide targeted recommendations for improving energy efficiency[17].

The VEU program has been successful in driving the adoption of energy-efficient technologies. Since the introduction of new electrification activities in mid-2023, over 24,000 Victorians have switched to efficient electric heating and cooling, and over 34,000 have upgraded to heat pump water heaters[18]. Importantly, 90% of VEU participants are satisfied with the program and the installed products, highlighting its positive impact on consumers[19].

### **Energy Efficiency in Social Housing Program (EESHP)**

The EESHP is a joint initiative between the Victorian and Australian governments to improve the energy efficiency and liveability of social housing properties[20]. The program aims to deliver up to 45,000 individual energy efficiency upgrades, including 5,000 fully electrified homes[21].

The EESHP specifically targets older social housing dwellings with gas appliances and low energy efficiency[22]. Renters in eligible properties are offered a package of upgrades to significantly improve their energy efficiency[23].

### **Comprehensive Retrofit Trial**

Sustainability Victoria conducted a Comprehensive Retrofit Trial to assess the potential for energy efficiency upgrades in existing homes[24]. The trial involved a range of building shell, appliance, and lighting upgrades in 60 houses representative of Victoria’s housing stock[25].

The trial demonstrated that significant energy and greenhouse gas savings could be achieved through comprehensive retrofits[26]. For example, the study found that comprehensive retrofits could reduce energy consumption by up to 70% and greenhouse gas emissions by up to 80%. The findings of this trial inform policy development and program design for improving energy efficiency in existing homes.

## **Electrification of Homes**

Victoria is actively pursuing the electrification of homes to reduce reliance on gas and transition to cleaner energy sources. This policy direction aligns with the Victorian government’s support for repowering homes with efficient electric appliances, a vital approach to lowering energy prices and reducing emissions[27].

### **Gas Substitution Roadmap**

The Gas Substitution Roadmap outlines Victoria’s plan to phase out gas connections in new homes and transition to all-electric appliances[28]. This roadmap supports the state’s goal of achieving net-zero emissions by 2045.

### **Regulatory Impact Statement on Building Electrification**

In 2024, the Victorian Government released a Regulatory Impact Statement on building electrification measures[29]. The preferred policy includes requiring gas appliances to be replaced with electric options at the end of their life and mandating all-electric systems in new homes and many new commercial buildings[30].

These measures are expected to generate significant bill savings for Victorian households and contribute to the state’s emissions reduction targets[31]. For example, data from Renew indicates that ditching gas and switching to reverse cycle split systems for heating can save at least two-thirds of annual energy bills, with the upfront investment paying for itself within one year[32].

## **Energy Efficiency Regulations for Rental Properties**

The Victorian Government is committed to improving energy efficiency in rental properties to ensure affordable and comfortable living conditions for tenants. It is important to note that energy efficiency in rental properties is often lower than in owner-occupied homes, highlighting the need for specific regulations and support in this area[33].

### **Minimum Energy Efficiency Standards**

Rental homes in Victoria are required to have a fixed heater in the main living area with at least a 2-star energy rating[34]. In December 2023, the Gas Substitution Roadmap update proposed new energy efficiency standards for rental properties, including ceiling insulation, draught sealing, efficient hot water systems and showerheads, and cooling[35].

## **Conclusion**

Victoria has implemented a comprehensive suite of policies, regulations, and schemes to improve energy efficiency in residential housing. The state’s commitment to net-zero emissions by 2045[36] is driving the transition to energy-efficient building practices, the adoption of renewable energy technologies, and the electrification of homes.

The VEU program, EESHP, and proposed minimum energy efficiency standards for rental properties are key initiatives that are transforming the residential housing sector in Victoria. These measures are contributing to lower energy bills for households[37], improved thermal comfort[38], and reduced greenhouse gas emissions[39].
