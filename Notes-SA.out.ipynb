In line with this ambition, the state government is actively promoting energy efficiency in residential housing through a range of policies, regulations, schemes, and other measures. This report provides a comprehensive overview of these efforts, drawing on a variety of sources, including government websites, academic research papers, reports from NGOs, news articles, and websites of energy providers.

## **Building Regulations**

The Building Code of Australia (BCA) is a national code adopted by each state and territory, setting minimum energy efficiency standards for new residential buildings. In South Australia, the BCA is administered by the Department for Infrastructure and Transport[9]. The BCA’s energy efficiency provisions aim to reduce energy consumption and greenhouse gas emissions from residential buildings by setting requirements for:

-   **Building envelope:** This includes requirements for insulation, glazing, and air tightness to minimize heat loss and gain.
-   **Heating and cooling systems:** This includes minimum energy efficiency standards for heating and cooling systems, such as air conditioners and heat pumps.
-   **Hot water systems:** This includes requirements for the energy efficiency of hot water systems, such as solar hot water systems and heat pumps.
-   **Lighting:** This includes requirements for energy-efficient lighting, such as LED lights.

## **Energy Efficiency Schemes and Incentives**

The South Australian government offers various schemes and incentives to encourage energy efficiency in residential housing:

-   **Retailer Energy Productivity Scheme (REPS):** This scheme requires energy retailers to meet energy savings targets by providing energy efficiency products and services to households[10]. Retailers can achieve these targets by offering discounts on energy-efficient appliances, providing energy audits, and installing energy-saving devices. To meet their targets, retailers can use energy credits earned in previous years, although the use of credits is being phased out[11].

-   **Home Energy Efficiency Scheme:** This scheme provides financial assistance to homeowners to install energy-efficient products and appliances[12]. Eligible upgrades may include insulation, solar hot water systems, and energy-efficient lighting. The scheme aims to reduce household energy consumption and lower energy bills.

-   **Community Batteries for Household Solar:** This program, funded by the Australian Renewable Energy Agency (ARENA), supports the installation of community batteries in South Australia[13]. These batteries allow households with solar panels to store excess energy and use it during peak demand periods, reducing reliance on the grid and maximizing the benefits of solar power. The program also provides eligible SA Housing Trust tenants with access to an exclusive energy plan with a reduced tariff[14].

## **Other Energy Efficiency Measures**

In addition to government schemes, homeowners can implement various measures to improve the energy efficiency of their homes:

-   **Installing insulation:** Insulation helps to reduce heat loss in winter and heat gain in summer, leading to lower energy consumption for heating and cooling[15]. Different types of insulation are available for ceilings, walls, and floors.

-   **Using energy-efficient appliances:** Energy-efficient appliances, such as refrigerators, washing machines, and dishwashers, consume less energy than standard appliances[16]. Look for appliances with high energy star ratings to maximize energy savings.

## **Role of the Essential Services Commission of South Australia**

The Essential Services Commission of South Australia (ESCOSA) plays a vital role in promoting energy efficiency through various mechanisms:

-   **Setting energy prices:** ESCOSA sets energy prices to ensure they are fair and reasonable, which can incentivize energy efficiency by making it more expensive to use excessive energy[17]. By reflecting the true cost of energy, ESCOSA’s pricing structures encourage consumers to reduce their energy consumption.

-   **Promoting competition in the energy market:** Competition in the energy market can lead to lower prices and more innovative energy efficiency products and services[18]. ESCOSA fosters competition by ensuring a level playing field for energy retailers and encouraging new entrants into the market.

## **Energy Providers**

Energy providers in South Australia also contribute to promoting energy efficiency. While specific programs and rebates vary between providers, some common offerings include:

-   **Discounts on energy-efficient appliances:** Some providers offer discounts on energy-efficient appliances, such as refrigerators and washing machines, to encourage customers to upgrade to more energy-saving models[19].

-   **Energy audits:** Some providers offer free or subsidized energy audits to help customers identify areas where they can improve energy efficiency in their homes[20].

-   **Installation of energy-saving devices:** Some providers offer free or subsidized installation of energy-saving devices, such as smart thermostats and energy-efficient lighting[21].

-   **Bundled discounts:** Dodo offers a monthly discount to customers who bundle their electricity and nbn plans[22].

## **NGOs and Other Organizations**

Several NGOs and other organizations are actively involved in promoting energy efficiency in South Australia:

-   **The Energy Efficiency Council (EEC):** The EEC is a not-for-profit membership association that promotes energy efficiency across the Australian economy[23]. They advocate for policies that support energy efficiency, provide information and resources to businesses and consumers, and promote best practices in energy management.

-   **CORENA South Australia:** CORENA is a not-for-profit organization that supports community-based renewable energy projects[24]. They provide funding and support for projects that promote the uptake of renewable energy and energy efficiency in communities.

-   **Tutti Arts South Australia:** Tutti Arts is a multi-arts organization that supports artists with disabilities[25]. They have a program that provides energy-efficient lighting to artists, helping them reduce their energy consumption and costs.

## **Academic Research**

Academic research provides valuable insights into the factors that influence energy consumption in homes and the effectiveness of different energy efficiency measures. Some key findings from research on energy efficiency in residential housing in South Australia include:

-   **Gap between expected and actual benefits:** Research has shown that there can be a gap between the expected and actual benefits of energy-efficient housing[26]. This highlights the importance of considering occupant behavior and providing ongoing support to ensure that energy-efficient features are used effectively.

-   **Incentives and barriers to adoption:** Financial incentives, such as rebates and grants, can play a significant role in encouraging the adoption of energy-efficient measures[27]. However, barriers such as upfront costs and lack of awareness can hinder the uptake of energy efficiency.

-   **Role of occupants:** Occupant behavior plays a crucial role in achieving energy savings in residential buildings[28]. Educating occupants about energy-efficient practices and providing feedback on their energy consumption can help to maximize the benefits of energy-efficient features.

## **Social Housing**

The South Australian government is investing in energy efficiency upgrades for social housing properties[29]. This initiative aims to improve the comfort and affordability of social housing for tenants while reducing energy consumption and greenhouse gas emissions. Upgrades may include insulation, energy-efficient appliances, and solar panels.

## **Support for Businesses**

Recognizing the impact of rising energy costs on small businesses, the South Australian government has increased its support for energy efficiency in the business sector[30]. Due to unprecedented demand for the Energy Efficiency Grants scheme, the government lifted the cap on Round 2 of the program, allowing more businesses to access grants for energy-efficient equipment and upgrades[31]. This demonstrates the government’s commitment to supporting businesses in reducing their energy costs and improving their sustainability.

## **Future Policy Directions**

South Australia is exploring new policy mechanisms to ensure grid stability and reliability as it transitions to a 100% renewable energy grid by 2027[32]. One such mechanism is the proposed “Firm Energy Reliability Mechanism,” which would provide payments to long-duration energy storage, including batteries and potentially gas generators[33]. This mechanism aims to address the challenges of “dunkelflaute,” a term describing periods of low wind and solar generation[34].

## **Energy Equity and Access**

Ensuring equitable access to energy efficiency measures is crucial for achieving a just energy transition. While South Australia has made significant progress in promoting energy efficiency, challenges remain in ensuring that all households, including those in remote communities, can benefit from these initiatives[35]. Targeted programs and support are needed to address the specific needs of remote and disadvantaged communities, ensuring that they can access the benefits of energy efficiency and contribute to a sustainable energy future.

## **Conclusion**

South Australia has a strong commitment to energy efficiency in residential housing, driven by its ambitious climate targets and a desire to reduce energy costs for households and businesses. The government has implemented a comprehensive suite of policies, regulations, schemes, and incentives to promote energy efficiency, complemented by the efforts of energy providers, NGOs, and academic researchers. While significant progress has been made, challenges remain in ensuring equitable access to energy efficiency measures and maintaining grid stability as the state transitions to a renewable energy future.
