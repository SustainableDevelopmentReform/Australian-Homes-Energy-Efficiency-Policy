## **Energy Efficiency Schemes and Programs**

### **Energy Ahead**

Formerly known as the Household Energy Efficiency Scheme (HEES), Energy Ahead is a vital program designed to assist WA households in enhancing the energy efficiency of their homes[1]. Delivered in partnership with Energy Policy WA, state-owned electricity retailers Synergy and Horizon Power, and non-government organizations (NGOs)[2], this program offers practical solutions and support to eligible households.

WA was once a leader in hardship energy efficiency programs, but the program was cut in 2012[3]. The reintroduction of a program with similar goals through Energy Ahead highlights a renewed focus on energy efficiency and its importance in addressing cost-of-living pressures and climate change. The collaborative approach involving NGOs, Synergy, Horizon Power, and Energy Policy WA ensures that the program can effectively identify and support households most in need[4].

Key features of Energy Ahead include:

-   **One-on-one coaching:** Eligible households receive personalized guidance from non-government organizations and retailers to identify and implement energy-saving measures. This coaching helps improve energy literacy and empowers residents to make informed decisions about their energy consumption[5]. Energy Ahead empowers households to understand their energy consumption and adopt energy-efficient behaviors and strategies[6].
-   **Appliance replacement:** In certain cases, the program facilitates the replacement of inefficient appliances, such as refrigerators, with more energy-efficient models, further contributing to energy savings and reduced household expenses[7].
-   **Financial assistance:** Energy Ahead aims to alleviate cost-of-living pressures for thousands of Western Australian households experiencing financial hardship by reducing energy bills through improved efficiency[8].

### **Nationwide House Energy Rating Scheme (NatHERS)**

While not a scheme in itself, NatHERS plays a crucial role in promoting energy efficiency in Australian homes, including those in WA. NatHERS provides a star rating system (0 to 10 stars) that indicates the energy efficiency of a home, with higher ratings signifying better performance. This rating system helps homeowners and builders understand the energy performance of their homes and make informed decisions about design and construction to optimize energy efficiency.

Interestingly, the average Australian household has an energy efficiency rating as low as 1.8 out of 10 on the NatHERS scale[9]. This finding has significant implications for WA’s residential sector, suggesting a considerable potential for improvement in energy efficiency across the state.

## **Energy Efficiency Regulations**

WA has implemented various regulations to ensure the safety and efficiency of energy systems in residential buildings. These regulations cover aspects such as electrical work, gas installations, and appliance standards. Key regulations include:

-   **Electricity (Licensing) Regulations 1991:** These regulations govern the standards for electrical work in residential installations, licensing of electricians and contractors, and related procedural matters[10]. They also cover the WA Electrical Requirements, which specify supply service rules. These rules dictate the technical requirements for connecting a property to the electricity grid, including voltage levels, metering, and safety standards.
-   **Gas Standards Regulations:** Several regulations under the Gas Standards Act 1972 address gas supply, system safety, gasfitting, consumer installations, and infringement notices, ensuring the safe and efficient use of gas in homes[11].
-   **Minimum Energy Performance Standards (MEPS):** While implemented at the national level, MEPS play a significant role in WA by setting minimum energy efficiency requirements for appliances and equipment. These standards ensure that new appliances sold in WA meet certain energy efficiency criteria, contributing to reduced energy consumption and lower household running costs[12].

## **National Context and Policy Recommendations**

The Australian Energy Efficiency Policy Handbook[13] provides a national context for energy efficiency policies and offers recommendations for improvement. It highlights the need for:

-   **Energy market reform:** This includes implementing fair and efficient electricity tariff structures, improving the economic efficiency of electricity networks, and ensuring that Network Service Providers invest in demand-side measures to reduce expenditure.
-   **Energy retailer obligations and energy efficiency funds:** The handbook recommends establishing energy efficiency funds in every jurisdiction and strengthening energy retailer obligations through energy efficiency schemes.
-   **Greenhouse gas reduction policies:** It emphasizes the importance of aligning and integrating energy market and climate change policies to drive energy efficiency.

## **Initiatives and Strategies**

### **Western Australian Climate Policy**

The WA Climate Policy outlines the state government’s commitment to addressing climate change and transitioning to a low-carbon future[14]. This policy recognizes the importance of balancing economic growth with reductions in greenhouse gas emissions, particularly in the context of renewable energy policies[15]. It includes several actions aimed at improving energy efficiency in the residential sector, such as:

-   **Energy efficiency in social housing:** The government is actively working to improve the energy efficiency of social housing, reducing energy costs for low-income households and contributing to emissions reduction targets[16].
-   **Investment in climate resilience:** The policy emphasizes investments in infrastructure to enhance climate resilience, including coastal adaptation and protection grants, which indirectly support energy efficiency by reducing the vulnerability of homes to extreme weather events[17].
-   **Electric vehicle (EV) charging infrastructure:** The policy supports the development of EV charging infrastructure, encouraging the uptake of EVs and contributing to a cleaner transport sector, which indirectly reduces reliance on fossil fuels for energy generation[18].

### **Energy Transformation Strategy**

Building upon the foundation of the WA Climate Policy, the WA Energy Transformation Strategy outlines a roadmap for the state’s energy future, focusing on a transition towards renewable energy sources and a more sustainable energy system[19]. This strategy includes initiatives that promote energy efficiency in residential housing, such as:

-   **Schools Virtual Power Plant and Smart Energy for Social Housing scheme:** These initiatives explore innovative approaches to energy management and efficiency in residential settings, potentially leading to wider adoption of similar technologies in the future[20].
-   **Pilot projects:** The strategy emphasizes the importance of pilot projects to test and challenge the existing power system, identifying opportunities to enhance energy generation, access, sharing, and consumption, which could lead to improved energy efficiency in homes[21].
-   **Energy sector governance reforms:** The strategy includes plans to investigate further governance changes to improve the efficiency of energy sector policy and decision-making, potentially leading to more effective energy efficiency regulations and programs[22].
-   **Technology upgrades:** Drawing insights from the mining sector, where companies like Wesfarmers Coal Limited have achieved significant energy savings through technology upgrades[23], the strategy recognizes the potential for similar improvements in the residential sector. This could involve promoting the adoption of energy-efficient appliances, smart home technologies, and renewable energy solutions in homes.

## **Case Studies and Examples**

While limited specific case studies were found focusing solely on WA residential energy efficiency policies, several examples highlight the implementation and impact of these measures:

-   **Energy Ahead case study:** A customer on a disability support pension with significant energy costs was referred to Energy Ahead. Through the program, they received assistance in understanding their energy consumption, implementing energy-saving strategies, and managing their energy bills, resulting in reduced financial hardship[24]. This case study demonstrates the program’s effectiveness in providing targeted support to vulnerable households and empowering them to improve their energy efficiency.

## **Challenges and Future Directions**

Despite the progress made in promoting energy efficiency in WA’s residential housing sector, challenges remain. One notable challenge is the increasing energy use for thermal comfort in new dwellings, despite meeting minimum energy efficiency standards[25]. This highlights the limitations of current policies that focus primarily on energy efficiency on a square meter basis rather than overall household energy use.

To address this challenge, future policy directions should consider a more holistic approach to energy efficiency, targeting household energy use as a whole. This could involve:

-   **Incentivizing smaller dwelling sizes:** As floor area is a key determinant of energy use[26], policies could encourage the construction of smaller, more energy-efficient homes.
-   **Promoting energy-conscious design and behavior:** Policies could encourage the adoption of passive design principles, which utilize natural elements like sunlight and ventilation to reduce reliance on artificial heating and cooling.
-   **Strengthening energy efficiency requirements:** Regularly reviewing and updating minimum energy performance standards for new buildings could ensure that they remain effective in driving down energy consumption.

## **Conclusion**

Western Australia is actively pursuing energy efficiency in the residential housing sector through a combination of schemes, regulations, and initiatives. Programs like Energy Ahead provide direct support to households, while regulations ensure the safety and efficiency of energy systems. The WA Climate Policy and Energy Transformation Strategy provide a framework for long-term improvements in energy efficiency, with a focus on reducing emissions and transitioning to a more sustainable energy future.

While these efforts are commendable, the analysis reveals a need for a more holistic approach to energy efficiency in residential housing. Future policies should move beyond simply setting minimum standards and consider the overall energy use of households. This could involve incentivizing smaller dwelling sizes, promoting energy-conscious design and behavior, and strengthening energy efficiency requirements. By addressing these challenges and adopting a more comprehensive approach, WA can further enhance energy efficiency in the residential sector, contributing to both cost savings for households and a more sustainable energy future for the state.
