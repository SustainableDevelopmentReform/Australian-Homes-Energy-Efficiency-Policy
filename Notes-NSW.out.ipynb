# Notes: NSW

### Policies

NSW Government policies provide frameworks for energy efficiency in the residential housing sector, setting targets, defining standards, and guiding program development.

#### NSW Energy Efficiency Action Plan

The NSW Energy Efficiency Action Plan[1] focuses on reducing household energy use and improving business energy productivity. The plan sets targets for annual energy savings of 16,000 gigawatt-hours by 2020 and aims to support 220,000 low-income households in reducing energy consumption.

#### NSW Government Resource Efficiency Policy (GREP)

The NSW Government Resource Efficiency Policy (GREP)[2] addresses government operating costs and resource efficiency through measures in energy, water, waste, and air emissions. Key energy measures include:

-   Energy savings targets across government sites
-   Minimum National Australian Built Environment Rating System (NABERS) Energy ratings for offices and data centers
-   Standards for new electrical appliances and equipment
-   Standards for new buildings and fit-outs
-   Solar target implementation
-   Fuel efficiency standards for new light vehicles
-   6% GreenPower purchase requirement

#### Australian Energy Efficiency Policy Handbook

The Australian Energy Efficiency Policy Handbook[3] outlines energy efficiency policies in Australia. It addresses energy retailer obligations, climate change policy alignment, and energy market policy coordination. The handbook discusses national residential energy efficiency disclosure schemes, rental property standards, and public housing upgrade programs.

### Regulations

NSW Government regulations set standards for energy performance in residential housing. These regulations establish requirements for new and existing homes[4].

#### Building Sustainability Index (BASIX)

The Building Sustainability Index (BASIX)[5] sets minimum requirements for energy efficiency, water efficiency, and thermal comfort in new homes and renovations. BASIX supplements the National Construction Code (NCC) by providing a framework for sustainable building practices in NSW. The index evaluates climate zone, building design, and material selection factors to address environmental impact.

#### Home Energy Efficiency Retrofits - Measurement of Energy Savings (HEER) Method Guide

The HEER Method Guide[6] provides methodology for measuring energy savings from residential building retrofits. It specifies requirements for calculating and verifying energy savings and defines roles for Accredited Certificate Providers (ACPs) and purchasers.

#### Home Energy Ratings Disclosure Framework

The Home Energy Ratings Disclosure Framework[7] aims to standardize energy efficiency information disclosure for residential buildings. The framework coordinates with existing disclosure schemes and promotes consistency across jurisdictions. It focuses on providing clear energy performance information to inform consumer decisions.

### Schemes

NSW Government schemes provide financial incentives and support for residential energy efficiency upgrades.

#### Energy Savings Scheme (ESS)

The Energy Savings Scheme (ESS)[8] creates financial incentives for energy savings activities through a market for energy savings certificates (ESCs). Electricity retailers must acquire specified numbers of ESCs annually. The scheme continues until a comparable national scheme is introduced.

#### NSW Consumer Energy Strategy 2024

The NSW Consumer Energy Strategy 2024[9] includes initiatives for:

-   Community incentives
-   Community action investment
-   Incentive scheme improvements
-   EV charging development
-   Renter energy cost reduction
-   Apartment solar investment support
-   Social housing energy bill reduction
-   Financial assistance for bill payment
-   Community battery storage support

#### Household Energy Upgrades Fund

The Australian Government has allocated \$1.3 billion for the Household Energy Upgrades Fund[10]. The Clean Energy Finance Corporation partners with financial institutions to provide loans for upgrades including battery-ready solar PV and energy-efficient appliances. The fund includes \$300 million for social housing property upgrades.

## Energy Efficiency Upgrades for Residential Housing

Energy efficiency upgrades in residential housing can reduce energy consumption and operating costs. Available upgrades target different building aspects:

### Appliances and Lighting

Energy-efficient appliances and lighting upgrades include[14]: \* Replacement of older appliances with higher energy star rated models \* LED lighting installation to replace incandescent bulbs

### Insulation and Draught Proofing

Building envelope improvements include[15]: \* Ceiling, wall, and floor insulation installation \* Window and door gap sealing

### Heating and Cooling Systems

HVAC system upgrades include[16]: \* Replacement of older systems with efficient models \* Installation of reverse-cycle air conditioners \* Heat pump hot water system implementation

### Solar Panels and Battery Storage

Renewable energy installations include[17]: \* Rooftop solar panel systems \* Battery storage systems \* Grid connection optimization
